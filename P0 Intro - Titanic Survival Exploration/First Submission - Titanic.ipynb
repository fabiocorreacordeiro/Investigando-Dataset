After I finished the first project to the course, I will use the decision tree algorithm to try to predict the outcome variable.
First I will training the algorithm using the data training.

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Import the train dataset to "data_train"
in_file_train = 'titanic_data.csv'
data_train = pd.read_csv(in_file_train)

# Split the outcome column ("Survived)
outcomes = data_train['Survived']
data_train = data_train.drop('Survived', axis = 1)
print(data_train.head())
# Use to train only numerical columns 
data_train = data_train[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
data_train['Age'] = data_train['Age'].fillna(0)

# Change "male" to 0 and "female" to 1 in "Sex" column
data_train['Sex'][data_train['Sex'] == 'male'] = 0
data_train['Sex'][data_train['Sex'] == 'female'] = 1

print (data_train.head())
```

```python
# Split the train dataset in train and test
X_train, X_test, y_train, y_test = train_test_split(data_train, outcomes, test_size=0.1)

# Training the classifier using Decision Tree Algorithm
clf = tree.DecisionTreeClassifier(max_depth=4)
clf = clf.fit(X_train, y_train)

# Predict using the test features and compare with the test Outcome
predictions = clf.predict(X_test)
print (accuracy_score(y_test, predictions))
```

Now I will predict the test dataset using the trained algorithm

```python
in_file_test = 'titanic_test.csv'
data_test = pd.read_csv(in_file_test)

# Use to test only numerical columns 
ID = data_test['PassengerId']
data_test = data_test[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
data_test['Age'] = data_test['Age'].fillna(0)
data_test['Fare'] = data_test['Fare'].fillna(0)

# Change "male" to 0 and "female" to 1 in "Sex" column
data_test['Sex'][data_test['Sex'] == 'male'] = 0
data_test['Sex'][data_test['Sex'] == 'female'] = 1

predictions_test = pd.DataFrame({'PassengerId': ID, 'Survived':clf.predict(data_test)})
predictions_test.to_csv('predictions_titanic.csv', index= False)
predictions_test
```
